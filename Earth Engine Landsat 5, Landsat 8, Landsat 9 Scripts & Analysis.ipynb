```python
import ee
import pandas as pd

# Authenticate and initialize the Earth Engine library
ee.Authenticate()
ee.Initialize()

csv_file_path = '/Users/rohan/fire_archive_M-C61_481023_filtered.csv'
coords_df = pd.read_csv(csv_file_path)

def get_ndvi(lat, lon, start_date, end_date):
    point = ee.Geometry.Point(lon, lat)
    landsat5 = ee.ImageCollection('LANDSAT/LT05/C01/T1_SR') \
        .filterBounds(point) \
        .filterDate(start_date, end_date)
    
    def calculate_ndvi(image):
        ndvi = image.normalizedDifference(['B4', 'B3']).rename('NDVI')
        return image.addBands(ndvi)
    
    landsat5_with_ndvi = landsat5.map(calculate_ndvi)
    
    median_image = landsat5_with_ndvi.median()
    
    ndvi_values = median_image.select('NDVI').reduceRegion(
        reducer=ee.Reducer.median(),
        geometry=point,
        scale=30
    )
    
    return ndvi_values.get('NDVI').getInfo()

start_date = '2008-01-01'
end_date = '2008-12-31'

ndvi_results = []

for index, row in coords_df.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    try:
        ndvi_value = get_ndvi(lat, lon, start_date, end_date)
    except Exception as e:
        ndvi_value = None
        print(f"Error processing point ({lat}, {lon}): {e}")
    ndvi_results.append(ndvi_value)

# Add NDVI results to the DataFrame
coords_df['NDVI'] = ndvi_results

# Save the results to a new CSV file
coords_df.to_csv('ndvi_results2008v2.csv', index=False)

# Print the resulting DataFrame
print(coords_df)
```

```python
import ee
import pandas as pd

# Authenticate and initialize the Earth Engine library
ee.Authenticate()
ee.Initialize()

# Load CSV file from your local filesystem
csv_file_path = '/Users/rohan/fire_archive_M-C61_481023_filtered.csv' 
coords_df = pd.read_csv(csv_file_path)

# Function to get NDVI for a given point using Landsat 8
def get_ndvi(lat, lon, start_date, end_date):
    point = ee.Geometry.Point(lon, lat)
    landsat8 = ee.ImageCollection('LANDSAT/LC08/C01/T1_SR') \
        .filterBounds(point) \
        .filterDate(start_date, end_date)
    
    # Calculate NDVI for each image in the collection
    def calculate_ndvi(image):
        ndvi = image.normalizedDifference(['B5', 'B4']).rename('NDVI')
        return image.addBands(ndvi)
    
    landsat8_with_ndvi = landsat8.map(calculate_ndvi)
    
    # Reduce the collection to a single median image
    median_image = landsat8_with_ndvi.median()
    
    # Reduce the region
    ndvi_values = median_image.select('NDVI').reduceRegion(
        reducer=ee.Reducer.median(),
        geometry=point,
        scale=30
    )
    
    return ndvi_values.get('NDVI').getInfo()

# Define the date range for the year 2014
start_date = '2014-01-01'
end_date = '2014-12-31'

# Initialize list to store NDVI results
ndvi_results = []

# Iterate through the coordinates and get NDVI
for index, row in coords_df.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    try:
        ndvi_value = get_ndvi(lat, lon, start_date, end_date)
    except Exception as e:
        ndvi_value = None
        print(f"Error processing point ({lat}, {lon}): {e}")
    ndvi_results.append(ndvi_value)

# Add NDVI results to the DataFrame
coords_df['NDVI'] = ndvi_results

# Save the results to a new CSV file
coords_df.to_csv('ndvi_results2014.csv', index=False)

# Print the resulting DataFrame
print(coords_df)
```
